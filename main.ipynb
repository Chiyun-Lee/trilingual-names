```python
import pandas as pd
import numpy as np
import random
```

```python
l_pinyin = pd.read_csv("pinyin.tsv", sep = "\t", header = None).fillna("").loc[:, 1:].map(lambda x: x.strip(" ")).values.flatten()
l_pinyin = [pinyin for pinyin in np.unique(l_pinyin) if pinyin not in [""]]

blacklisted_substrings = [
    "x", "z", "uai", "er", "ao", "c", "q", "ou", "ui", "f", "yin", "sh",
    "Ã¼", "ie", "ai", "en", "ia", "iu", "ua", "un", "yan", "uo", "si", "shi"
]
blacklisted_prefixes = ["r"]
blacklisted_suffixes = ["e", "in"]

l_pinyin = [pinyin for pinyin in l_pinyin if not any (substring in pinyin for substring in blacklisted_substrings)]
l_pinyin = [pinyin for pinyin in l_pinyin if not any (pinyin.startswith(prefix) for prefix in blacklisted_prefixes)]
l_pinyin = [pinyin for pinyin in l_pinyin if not any (pinyin.endswith(  suffix) for suffix in blacklisted_suffixes)]

print(len(l_pinyin))

print(", ".join(l_pinyin))

l_pinyin
```

```python
manual_blacklist = [
    "ang", "bang", "bing", "dang", "bo", "dei", "ei", "ding", "dong",
    "gang", "gong", "gei", "jiong", "yo", "wei", "pei", "pang", "ping",
    "nang", "mang", "lei", "kong", "kang", "bei", "hang", "sang", "pi",
    "po", "nei", "ling", "ming", "hei", "gu", "lang", "wang", "ting", "tong",
    "tang", "hong", "jing", "nong", "long", "yong", "ba", "ju", "pan", "gan",
    "mo", "pu", "wo", "ga", "pa", "lo", "mei", "nu", "o", "wu", "tan", "bu",
    "nan", "man", "san", "kan", "lan", "wan", "ma", "tu", "ta", "ban", "song",
    "ning", "di"
]

l_curated = [p for p in l_pinyin if p not in manual_blacklist]
print(len(l_curated))
l_curated
```

```python
pd.set_option("display.max.rows", None)
query_pinyin = ""

if query_pinyin == "":
    query_pinyin = random.choice(l_curated)

names = [f"Li {query_pinyin.capitalize()}{p}; {query_pinyin.capitalize()}{p} Lee; Li {p.capitalize()}{query_pinyin}; {p.capitalize()}{query_pinyin} Lee" for p in l_curated]

print(query_pinyin)
pd.DataFrame([l.split("; ") for l in names])
```
